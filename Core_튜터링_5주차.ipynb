```
# 벽은 #로, 빈 공간은 .으로, 캐릭터는 B로 둔다.
Block = '#'
Void_area = '.'
Character = 'B'

# 게임판을 새롭게 정의
game_board = []
row = 0          # 행의 길이
column = 0       # 열의 길이

# 캐릭터의 자리 또한 새롭게 정의
c_y = 0
c_x = 0
```

함수 정의부

```
# 게임판을 초기화하는 함수
def default_game_board(row, column, c_y, c_x):
  game_board = []
  for i in range(0, row, 1):
    a = []
    for j in range(0, column, 1):
      a.append('.')
    game_board.append(a)
  
  #캐릭터의 자리 정하는 부분
  game_board[c_y][c_x] = 'B'

  #벽을 넣는 부분
  for i in range(0, column, 1):
    game_board[0][i] = '#'
    game_board[row-1][i] = '#'

  for i in range(0, row, 1):
    game_board[i][0] ='#'
    game_board[i][column-1] = '#'

  print_game_board(game_board)

  return game_board
```

```
# 이중 리스트를 이중 리스트답게 출력하는 함수를 만들어 보자
def print_game_board(game_board):
  print("이중 리스트를 출력합니다.")
  print(type(game_board), game_board)
  for i in range(0, len(game_board), 1): # 가로 (행)
    for j in range(0, len(game_board[i]), 1): # 세로 (열)
      print(game_board[i][j], end = " ")
    print()
  print("이중 리스트의 출력이 끝났습니다.")
  print()
```

```
# 테두리에 닿았다면?
# 내가 왼쪽 끝까지 왔는데, 또 왼쪽으로 간다고 하면
# 못가게 막는 것!

def check(dir, y, x):
  # 현재 좌표가 테두리를 벗어난다면
  if x < 0 or x >= 5 or y < 0 or y >= 5:
    
    # 다시 고쳐주고
    if dir == "왼쪽":
      x = 0
    elif dir == "오른쪽":
      x = 4
    elif dir == "위쪽":
      y = 0
    elif dir == "아래쪽":
      y = 4

    # 사용자에게 알려줍니다.
    print("더 이상 ", dir, "으로 갈 수 없습니다!")

  # 그리고 그 좌표를 리턴합니다.
  return y, x
```

```
# 재귀함수를 이용 -> 벽을 만나기 전까지 한 방향으로 계속 캐리거를 움직이는 함수
def move_not_block(game_board, c_y, c_x, dir):
  print_game_board(game_board)
  print(c_y, c_x, dir)

  # 함수 안에서만 쓰이는 y, x 변수를 만들자
  y = c_y
  x = c_x

  if dir == '왼쪽':
    c_x = c_x -1
  elif dir == '오른쪽':
    c_x = c_x + 1
  elif dir == '위쪽':
    c_y = c_y - 1
  elif dir == '아래쪽':
    c_y = c_y + 1

  if game_board[c_y][c_x] != '#':
    game_board[y][x] = '.'
    game_board[c_y][c_x] = 'B'
    return move_not_block(game_board, c_y, c_x, dir)
  else:
    return game_board
```
